```python
from azureml import Workspace

ws = Workspace()
ds = ws.datasets['apy.csv']
frame = ds.to_dataframe()
```

```python
frame.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
plt.hist(frame.groupby(by='State_Name')['Area'].mean())
plt.xlabel('average Area of crop grown in various states ')
```

```python
import numpy as np
```

```python
grp=pd.DataFrame(frame.groupby(by='Crop')['Area'].mean(),index=None)
```

```python
grp.reset_index(inplace=True)
```

```python
grp
```

```python
grp.Area=grp.Area/grp.Area.max()
```

```python
plt.figure(figsize=(20,10))
sns.barplot('Crop','Area',data=grp)
plt.xticks(rotation=90)
```

```python
sea=pd.DataFrame(frame.groupby(by='Season').mean()).reset_index()
sea
```

```python
#plt.figure(figsize=(14,10))
sns.barplot('Season','Production',data=sea,hue=sea.Area)
```

```python
crop_year=pd.DataFrame(frame.groupby('Crop_Year').mean()).reset_index()
crop_year
```

```python
sns.barplot('Crop_Year','Production',data=crop_year)
plt.xticks(rotation=45)
```

```python
dist_name=pd.DataFrame(frame.groupby('District_Name').mean()).reset_index()
dist_name.sort_values('Area',ascending=False)
```

```python
plt.plot('Area','Production','g--',data=dist_name)
```
